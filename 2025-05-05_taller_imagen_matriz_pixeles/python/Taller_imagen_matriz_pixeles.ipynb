```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Cargar imagen en color
image = cv2.imread('image.png')  # Asegúrate de que 'image.png' esté cargada en tu Colab
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convertir de BGR a RGB
```

```python
# Separar canales RGB
R, G, B = cv2.split(image_rgb)

# Convertir a HSV y separar canales
image_hsv = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2HSV)
H, S, V = cv2.split(image_hsv)

# Visualización
fig, axs = plt.subplots(2, 3, figsize=(15, 8))
axs[0, 0].imshow(R, cmap='Reds'); axs[0, 0].set_title("Canal R")
axs[0, 1].imshow(G, cmap='Greens'); axs[0, 1].set_title("Canal G")
axs[0, 2].imshow(B, cmap='Blues'); axs[0, 2].set_title("Canal B")
axs[1, 0].imshow(H, cmap='hsv'); axs[1, 0].set_title("Canal H")
axs[1, 1].imshow(S, cmap='gray'); axs[1, 1].set_title("Canal S")
axs[1, 2].imshow(V, cmap='gray'); axs[1, 2].set_title("Canal V")
for ax in axs.flatten(): ax.axis('off')
plt.tight_layout()
plt.show()
```

```python
modified = image_rgb.copy()
modified[50:150, 100:200] = [255, 0, 0]  # Área en rojo (rectángulo)
plt.imshow(modified)
plt.title("Región coloreada")
plt.axis('off')
plt.show()
```

```python
image_copy = image_rgb.copy()
h, w, _ = image_rgb.shape
image_copy[0:100, 0:100] = image_rgb[h-100:h, w-100:w]  # Copiar esquina inferior derecha a esquina superior izquierda
plt.imshow(image_copy)
plt.title("Región sustituida")
plt.axis('off')
plt.show()
```

```python
# Histograma del canal de intensidad en escala de grises
gray = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)
plt.hist(gray.ravel(), bins=256, range=[0, 256], color='black')
plt.title("Histograma de Intensidad (Grises)")
plt.xlabel("Intensidad")
plt.ylabel("Frecuencia")
plt.show()
```

```python
alpha = 1  # Contraste (>1 aumenta, <1 disminuye)
beta = -120    # Brillo (>0 aumenta, <0 disminuye)
adjusted_cv = cv2.convertScaleAbs(image_rgb, alpha=alpha, beta=beta)
plt.imshow(adjusted_cv)
plt.title("Ajuste con OpenCV")
plt.axis('off')
plt.show()
```
